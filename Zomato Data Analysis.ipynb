# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Create the data frame

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

# convert the data type of coloum - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value=value[0]
    return float(value)

df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

# Type of resturant

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlable("type of resturant")
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()

# Creating a DataFrame from the grouped data
result = pd.DataFrame({'vote': grouped_data})

# Plotting
plt.plot(result.index, result['vote'], c="green", marker="o")  
plt.xlabel("Type of Restaurant", color="red", size=20)
plt.ylabel("Votes", color="red", size=20)
plt.title("Votes by Type of Restaurant", size=22)
plt.xticks(rotation=45)  
plt.show()
```

```python
plt.hist(df['rate'],bins =5)
plt.title("ratings distribution")
plt.show()
```

# Average order spending by couples

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# Which mode receives maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order' , y = 'rate', data = df)
```

```python
# Creating a pivot table
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

# Plotting the heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap of Online Orders by Restaurant Type")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
